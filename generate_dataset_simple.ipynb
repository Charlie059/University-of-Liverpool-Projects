```

import numpy as np
import random
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
drive.mount('drive')

def generate_dataset_simple(beta, n, std_dev):
  x = np.random.random_sample(n) * 100
  b = np.random.randn()
  e = np.random.randn(n) * std_dev
  y = x * beta + b + e
  return x,y
  
def linear_regression_demo(n = 200,k=10,std_dev = 500):
  for i in range(10):
    

    x, y = generate_dataset_simple(10, n, std_dev)
    true_y = [ item * k for item in x]
    model = LinearRegression()
    model.fit(np.reshape(x,[len(x),1]), np.reshape(y,[len(y),1]))
    yy = model.predict(np.reshape(x,[len(x),1]))
    plt.figure()
    kk = model.coef_[0][0] 
    bb = model.intercept_[0] 
    plt.title('Linear Regression Demo \n True: y='+str(k)[0:4]+'x +'+str(b)[0:4]+'  Predicted:y='+str(kk)[0:4]+'x +'+str(bb)[0:4] );
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True) 
    plt.plot(x,y,'r.') 
    plt.plot(x,yy,'g-') 
    plt.show() 
    
    x = x.reshape(x.shape[0],)
    # initialise data of lists. 
    data = {'X':x,'y':y} 
    # Creates pandas DataFrame. 
    df = pd.DataFrame(data) 

    df.to_csv("/content/drive/MyDrive/C++ C2/Test_Data"+str(i)+".csv", index=False)



linear_regression_demo()


  

```
